```python
from janome.tokenizer import Tokenizer
from gensim.models import word2vec
import re # 正規表現を利用
```

```python
binarydata = open('01.data/ningen_shikkaku.txt', 'rb').read()
```

```python
# データ型を文字列型に変換
text = binarydata.decode('shift_jis')
```

```python
# 文章から余計なものを削除
text = re.split(r'\-{5,}', text)[2] # ハイフンが５つ以上並んでいる箇所で分割
text = re.split(r'底本：', text)[0] # フッターを削除
text = text.strip() # 余計なスペースを削除
```

## 形態素解析

```python
t =Tokenizer()
```

```python
lines = text.split('\r\n') # 文章を改行コードで分割
```

```python
results = []
for line in lines:
    s = line
    s =s.replace('|', '')
    s = re.sub(r"《.+?》","", s ) # ルビをとる   ［＃
    s = re.sub(r'［＃.+?］','', s ) # 入力注をとる
    
    tokens = t.tokenize(s)
    
    r =[]
    for token in tokens:
        if token.base_form == '*':
            w = token.surfce
        else :
            w =token.base_form
        
        ps = token.part_of_speech
        hinshi = ps.split(',')[0] # 複数の品詞の候補がある場合に、先頭のものを採用する
        if hinshi in ['名詞', '形容詞', '動詞', '記号']:
            r.append(w)
    
    rl = (' '.join(r)).strip()
    results.append(rl)
    print(rl)
```

```python
wakachi_file = '02.wakachi/ningen_shikkaku.wakachi'
with open(wakachi_file, 'w', encoding='utf-8') as fp : # 書き込み可能にして、utf8でfile pointerを定義
    fp.write('\n'.join(results))
```

## word2vec

```python
# モデルを作成
data = word2vec.LineSentence(wakachi_file)
model = word2vec.Word2Vec(data, size=200, window=10, hs =1, min_count=2, sg=1)
model.save('03.model_word2vec/ningen_shikkaku.model')
```

```python
model.most_similar(positive=['私'])
```
